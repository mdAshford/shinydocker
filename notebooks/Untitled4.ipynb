```python
n=10000
ρc = 322 #, "kg/m**3")
Tc = 647.096 #, "K")
To = 273.15 #, "K")

T = np.linspace( To, Tc, n)
vf = np.zeros(n, dtype=object)
vg = np.zeros(n, dtype=object)


```

```python
def v( T ):
    # T = T * 1u"K"
#     Tc = 647.096 #, "K") #u"K"
#     Pc = 22.064 #, "MPa") #u"MPa"
#     ρc = 322 #, "kg/m**3") #u"kg/m**3"
    b = [0, 1.99274064, 1.09965342, -0.510839303, -1.75493479, -45.5170352, -6.74694450e5]
    c = [0, -2.03150240, -2.68302940, -5.38626492, -17.2991605, -44.7586581, -63.9201063]
    ν = 1 - T/Tc
    ρf = ρc * (1 + b[1]*ν**(1/3) + b[2]*ν**(2/3) + b[3]*ν**(5/3) + b[4]*ν**(16/3) + b[5]*ν**(43/3) + b[6]*ν**(110/3) )
    ρg = ρc * math.exp( c[1]*ν**(2/6) + c[2]*ν**(4/6) + c[3]*ν**(8/6) + c[4]*ν**(18/6) + c[5]*ν**(37/6) + c[6]*ν**(71/6) )
    return [ 1/ρf, 1/ρg ]
```

```python
# n=10000
# ρc = 322 #, "kg/m**3")
# Tc = 647.096 #, "K")
# To = 273.15 #, "K")

# T = np.linspace( To, Tc, n)
# vf = np.zeros(n, dtype=object)
# vg = np.zeros(n, dtype=object)




# T = np.arange(n)/(n-1)*(Tc - To) + To
# T

for k in np.arange(n):
    vf[k] = v(T[k])[0]
    vg[k] = v(T[k])[1]

v = np.zeros(2*n, dtype=object)
v = np.append(vf,vg)

```

```python
# def plotthis(f):
#     print(x)
A = np.int(n/2)
x = 50
# A = NS

xfrac = (x/100)*math.log(vg[A]/vf[A],10)
vx = vf[A]*10**xfrac

p = figure(plot_width=900, plot_height=400, x_axis_type="log")
r = p.multi_line([vf, vg], [T, T], color=["blue", "blue"], alpha=[1, 1], line_width=2)

# sat1 = p.circle([vf[ns], vg[ns]], [T[ns], T[ns]], fill_color="red", line_color="red", size=6)
# sat2 = p.line([vf[ns], vg[ns]], [T[ns], T[ns]], line_color="red")

# qx = p.circle([(vf[ns]+vg[ns])/2], [T[ns]], size=9)


sat1 = p.circle([vf[A], vg[A]], [T[A], T[A]], fill_color="red", line_color="red", size=6)
sat2 = p.line([vf[A], vg[A]], [T[A], T[A]], line_color="red")

qx = p.circle([(vf[A]+vg[A])/2], [T[A]], size=9)


```

```python
def update(A=np.int(n/2), x=50):
#     nindex = n*A/100
    print(A)
    
    xfrac = (x/100)*math.log(vg[A]/vf[A],10)
    vx = vf[A]*10**xfrac
    
    sat1.data_source.data['y'] = [T[A], T[A]]
    sat1.data_source.data['x'] = [vf[A], vg[A]]

    sat2.data_source.data['y'] = [T[A], T[A]]
    sat2.data_source.data['x'] = [vf[A], vg[A]]

    qx.data_source.data['y'] = [T[A]]
    qx.data_source.data['x'] = [vx]   

    push_notebook()
```
