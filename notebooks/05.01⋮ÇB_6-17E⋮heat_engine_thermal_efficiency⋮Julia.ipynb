**6-17E.** A heat engine has a heat input of 3 × 10⁴ Btu/h and a thermal efficiency of 40 percent. Calculate the power it will produce, in hp.


![Cengelboles7ed4-126](./figures/ch06_cen98179_p06017_lg.jpg)


### Given
<ol> 
 $\dot Q_H = 30000\, \ce{Btu/hr}, \quad \eta_{th} = 40\% $    

### Find
<ol>
  
$\dot W_{net}$     


### Get Organized  and Attack 
<ol>
    
This is the textbook definition of thermal efficiency.

$$ \eta_{th} = \frac{\text{desired output}}{\text{required input}} = \frac{\dot W_{net}}{\dot Q_H} = 40\% \\[18pt] $$

$$ \dot W_{net} = 0.4\,\dot Q_H = 12000\, \rm \frac{Btu}{hr} \, \frac{hp}{2545\, Btu/hr}$$ 

<div class="panel-success"><div class="panel-heading"><div class="badge"></div> $$\ \normalsize \dot W_{net} = 4.715\, \ce{hp}$$
</div></div>
